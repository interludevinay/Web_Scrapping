```python
soup = BeautifulSoup(requests.get("https://www.cars24.com/buy-used-car?sort=bestmatch&serveWarrantyCount=true&storeCityId=2378&pinId=400001").text)
soup
```

```python
name = []
detail = []
km = []
owner =[]
type = []
for i in soup.find_all('div',{'class':'_2z-Yu'}):
  for t in soup.find_all('div',{'class':'_1mlAq'}):
    for y in soup.find_all('h2',{'class':'_2lmIw'}):
      name.append(t.text)
      for u in soup.find_all('ul',{'class':'_13yb6'}):
        detail.append(u.text)

detail = detail[1:1000]
name = name[1:1000]
print(detail)
for j in detail:
  k = j[0:7]
  km.append(k)
  l = j[9:18]
  owner.append(l)
  o = j[18:24]
  type.append(o)
print(km)
print(owner)
print(type)

```
